# Clustering

## Mencari nilai optimum K

Disini saya menggunakan 2 variabel, yaitu SCG dan PEG. Pemiilihan ini dilakukan untuk mencari seberapa berpengaruhnya pengulangan materi terhadap hasil ujian

```python
X = user[['SCG' , 'PEG']].iloc[: , :].values
inertia = []
for n in range(1 , 10):
    models = (KMeans(n_clusters = n ,init='k-means++', n_init = 10 ,max_iter=100, 
                        tol=0.0001,  random_state= 100  , algorithm='elkan') )
    models.fit(X)
    inertia.append(models.inertia_)
```

## Elbow Method

```python
plt.figure(1 , figsize = (15 ,6))
plt.plot(np.arange(1 , 10) , inertia , 'o')
plt.plot(np.arange(1 , 10) , inertia , '-' , alpha = 0.5)
plt.xlabel('Number of Clusters') , plt.ylabel('Inertia')
plt.show()
```

Dari gambar diatas, diambil kesimpulan bahwa nilai 3 adalah yang optimum untuk digunakan karena itu adalah titik dimana penurunan nilai WCSS (Within-Cluster Sum of Square) mulai melambat dan mendekati garis lurus. Disini terlihat bahwa titik 2 adalah titik dimana WCSS menurun paling tajam. Namun, titik ini tidak selalu merupakan pilihan terbaik untuk k, karena bisa jadi terlalu sedikit cluster untuk menangkap variasi data.

```python
models = (KMeans(n_clusters = 3 ,init='k-means++', n_init = 10 ,max_iter=300, 
                        tol=0.0001,  random_state= 111  , algorithm='elkan') )
models.fit(X)
labels = models.labels_
centroids = models.cluster_centers_
```

```python
print(models.cluster_centers_)
print(models.inertia_)
print(models.n_iter_)
```

Nilai inersia dari model ini adalah 9,16. Inersia adalah jumlah kuadrat jarak sampel ke pusat cluster terdekat.
Konvergensi merujuk pada tahap di mana model telah mencapai kestabilan atau solusi yang optimal. Dari data diatas, untuk mencapai konvergensi, model ini membutuhkan 5 kali iterasi

```python
fig = plt.figure(figsize=(12,8))
plt.scatter(X[:,0], X[:,1], c=models.labels_, cmap="Set1_r", s=25)
plt.scatter(models.cluster_centers_[:,0] ,models.cluster_centers_[:,1], color='blue', marker="*", s=250)
plt.title("Persebaran Data Cluster", fontsize=16)
plt.xlabel("Pengulangan Materi", fontsize=16)
plt.ylabel("Nilai Ujian", fontsize=16)
plt.show()
```
